```python
G = cv2.imread(r"C:\Users\soura\Documents\Sem 6 Labs\IPMV\Images\cameraman.tif")
G = cv2.cvtColor(G, cv2.COLOR_BGR2RGB)
plt.subplot(2,2,1)
plt.imshow(G, cmap='gray')
G = G.astype(float)

Gmin = np.min(G)
Gmax = np.max(G)

Gnorm = ((G - Gmin) / (Gmax - Gmin)) * 256
Gnorm = Gnorm.astype(np.uint8)

plt.subplot(2,2,2)
plt.hist(G.ravel(), bins=256, range=(0,256))

plt.subplot(2,2,3)
plt.imshow(Gnorm, cmap='gray')

plt.subplot(2,2,4)
plt.hist(Gnorm.ravel(), bins=256, range=(0,256))

plt.show()
```
