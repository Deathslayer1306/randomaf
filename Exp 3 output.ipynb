```python
#pip install pandas numpy opencv-python matplotlib  
import   pandas   as   pd  
import   numpy   as   np    
import   cv2  
import   matplotlib.pyplot   as   plt  
import   math  
img   =   cv2 . imread ( r "C:\Users\admin\Documents\IPMV\EXP 3\noisy_image\watermark.jpg" )   
img   =   cv2 . cvtColor ( img ,   cv2 . COLOR_RGB2BGR )   
plt . figure ( figsize = ( 5 , 4 ))    
plt . imshow ( img )  
plt . axis ( "off" )  
plt . show ()  
#---------Salt and pepper -------  
def   salt_pepper ( img ,   prob   =   0.08 ):       
    noisy   = img . copy ()       
    h , w   =   img . shape [: 2 ]       
    for   i   in   range ( h ):           
        for   j   in   range ( w ):               
            r   =   np . random . rand ()               
            if   r   <   prob / 2 :                   
                noisy [ i , j ]   =   0               
            elif   r   <   prob :                   
                noisy [ i ,   j ]   =   255       
    return   noisy 
sp   =   salt_pepper ( img )
plt . figure ( figsize = ( 10 , 8 ))   
images   =   [ img ,   sp ]  
titles   =   [ "Original" ,   " Salt & Pepper" ]    
for   i   in   range ( 2 ):        
    plt . subplot ( 2 , 2 , i + 1 )        
    plt . imshow ( images [ i ])        
    plt . title ( titles [ i ])        
    plt . axis ( "off" )   
plt . show ()
```

```python
h , w   =   img . shape [: 2 ]  
img . shape [: 2 ]  
img . shape  (437, 589, 3)  


def   gaussian_noise ( img ,   sigma   =   50 ):       
    guass   =   np . random . normal ( 0 ,   sigma ,   img . shape )       
    noisy   =   img   +   guass      
    noisy   =   np . clip ( noisy ,   0 ,   255 )       
    return   noisy . astype ( np . uint8 )  
ga   =   gaussian_noise ( img )  
images   =   [ img ,   ga ]  
titles   =   [ "Original" ,   " Gaussian" ]    
for   i   in   range ( 2 ):        
    plt . subplot ( 2 , 2 , i + 1 )        
    plt . imshow ( images [ i ])        
    plt . title ( titles [ i ])        
    plt . axis ( "off" )   
plt . show ()
```

```python
def   blur_noise ( img ,   k = 15 ):       
    return   cv2 . GaussianBlur ( img ,   ( k ,   k ),   0 )  
bl   =   blur_noise ( img ,   k = 15 )  
images   =   [ img ,   bl ]  
titles   =   [ "Original" ,   " Blurred" ]    
for   i   in   range ( 2 ):        
    plt . subplot ( 2 , 2 , i + 1 )        
    plt . imshow ( images [ i ])        
    plt . title ( titles [ i ])        
    plt . axis ( "off" )   
    plt . show ()  


def   mse ( original ,   noisy ):        
    return   np . mean (( original   -   noisy )   **   2 )    
def   psnr ( original ,   noisy ):        
    m   =   mse ( original ,   noisy )       
    if   m   ==   0   :           
        return   100       
    return   20   *   math . log10 ( 255   /   math . sqrt ( m ))  


print ( " \n Image Quality Metrics \n " )  
print ( "Salt & Pepper" )  
print ( "MSE:" ,   mse ( img ,   sp ))  
print ( "PSNR:" ,   psnr ( img ,   sp ),   " \n " )   
print ( "Gaussian Noise" )  
print ( "MSE:" ,   mse ( img ,   ga ))  
print ( "PSNR:" ,   psnr ( img ,   ga ),   " \n " )
```
