```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import math
```

```python
img = cv2.imread(r"C:\Users\soura\Documents\Sem 6 Labs\IPMV\Images\lena.tif")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(5,5))
plt.imshow(img)
```

```python
salt_pepper = cv2.imread(r"C:\Users\soura\Documents\Sem 6 Labs\IPMV\output\sp_image.png")
gaussian = cv2.imread(r"C:\Users\soura\Documents\Sem 6 Labs\IPMV\output\gn_image.png")
blurred = cv2.imread(r"C:\Users\soura\Documents\Sem 6 Labs\IPMV\output\blr_image.png")


median_restored = cv2.medianBlur(salt_pepper, 5)
plt.figure(figsize=(4,4))
plt.imshow(median_restored)

gauss_restored = cv2.GaussianBlur(gaussian,(5,5),0)
plt.figure(figsize=(4,4))
plt.imshow(gauss_restored)

blur_restored = cv2.blur(blurred,(5,5),0)
plt.figure(figsize=(4,4))
plt.imshow(blur_restored)
```

```python
def mse(original, noisy):
    return np.mean((original-noisy)**2)
def psnr(original, noisy):
    m = mse(original, noisy)
    if m == 0:
        return float('inf')
    else:
        return 20* math.log10(255/math.sqrt(m))

print("MSE=", mse(img, median_restored))
print("PSNR=", psnr(img, median_restored))
```
